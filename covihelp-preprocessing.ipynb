```python
def datetimeparser(df):
    dates = df.tolist()
    new_dates=[]
    for date in dates:
        try:
            new_dates.append(pd.to_datetime(date, format='%d/%m/%Y %H:%M:%S'))
        except ValueError:
            try:
                new_dates.append(pd.to_datetime(date, format='%H:%M-%d/%m/%Y'))
            except ValueError:
                try:
                    new_dates.append(pd.to_datetime(date, format='%d/%m/%Y-%H:%M'))
                except ValueError:
                    try:
                        new_dates.append(pd.to_datetime(date, format='%d/%m/%Y'))
                    except:
                        try:
                            new_dates.append(pd.to_datetime(date, format='%d/%m/%Y %H:%M'))
                        except:
                            try:
                                new_dates.append(pd.to_datetime(date, format='%m/%d/%Y %H:%M:%S'))
                            except:
                                new_dates.append(pd.to_datetime(date, format='%m/%d/%Y %H:%M'))
    return new_dates
```

```python
df3['Timestamp'] = datetimeparser(df3['Timestamp'])
df2['Timestamp'] = datetimeparser(df2['Timestamp'])
df1['Timestamp'] = datetimeparser(df1['Timestamp'])
```
